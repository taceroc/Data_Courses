```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
%matplotlib inline  
```

```python
url = "http://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df = pd.read_csv(url, header = None) 
headers = ["symboling","normalized-losses","make","fue-type","aspiration","num-of-doors"," body-style",
           "drive-wheels","engine-location","wheel-base","length","width","heigh","curb-weigh","engine-type",
           "num-of-cylinders","engine-size","fuel-system","bore","stroke","compression-ratio","horsepower","peak-rpm",
          "city-mpg","highway-mpg","price"]
df.columns = headers
df.head() #Show first n row of data frame fila xd
#df.tail() #Show bottom n row of data frame
```

```python
#df["city-mpg"] = 235/df["city-mpg"] #transform mpg to L/100km
#df.rename(columns={"city-mpg":"city-L/100km"},inplace=True)
df["city-L/100km"] = 235/df["city-mpg"] #transform mpg to L/100km
```

```python
df.head()#df["city-L/100km"]
```

```python
df.replace('?', np.nan, inplace=True)
df['normalized-losses'] = df['normalized-losses'].astype('float')
```

```python
df["normalized-losses"].dtypes
```

# Normalization

```python
df["1city-L/100km"] = df["city-L/100km"]/df["city-L/100km"].max()
```

```python
df["1city-L/100km"]
```

```python
df["2city-L/100km"] = (df["city-L/100km"]-df["city-L/100km"].min())/(df["city-L/100km"].max()-df["city-L/100km"].min())
```

```python
df["2city-L/100km"]
```

```python
df["3city-L/100km"] = (df["city-L/100km"]-df["city-L/100km"].mean())/(df["city-L/100km"].std())
```

```python
df["3city-L/100km"]
```

```python
df
```

# Binning

```python
df["price"].dtypes
```

```python
df["price"] = df["price"].astype('float')
```

```python
df["price"].dtypes
```

```python
bins = np.linspace(min(df['price']),max(df['price']),4)
```

```python
group_names = {"Low","Medium","High"}
```

```python
df["price binned"] = pd.cut(df['price'], bins, labels=group_names,include_lowest=True)
```
